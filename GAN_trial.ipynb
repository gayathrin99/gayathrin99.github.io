```python
def load_images(directory, n):
  images = []
  i=0
  for file in listdir(directory):
    filename= directory + file
    image = cv2.imread(filename)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    image = np.asarray(image)
    images.append(image)
    i += 1
    if i>= n:
      break
      return np.asarray(images)
```
